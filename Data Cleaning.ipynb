### Transform Data

We need to transform dataset to aggregate the number of trips for each station every hour, and determine
- Average trip_duration
- Count number of user_types
- Count number of genders
- is_weekend, is_holiday columns
- weather data

```python
df_bike_transform = df_bike_clean
df_bike_transform.head()
```

#### Transform start_time: Round to lowest hour

```python
df_bike_transform['start_time'] = df_bike_transform['start_time'].dt.floor('60min')
```

```python
df_bike_transform.head()
```

#### Group data by start_time and start_station_id
- Count total trips
- Average trip_duration
- Count number of user_types
- Count number of genders
- is_weekday

##### Total trips, average trip duration

```python
# Total trips, average trip duration for each group of start_time/start_station_id
df_group_trip_duration = df_bike_transform.groupby(['start_time', 'start_station_id']).agg({'trip_duration': ['count','mean']})
```

##### Number of each user_type, gender

```python
# Number of each gender/user_type for each group of start_time/start_station_id
df_group_genders = df_bike_transform.groupby(['start_time', 'start_station_id']).gender.value_counts().unstack(fill_value=0)
df_group_user_type = df_bike_transform.groupby(['start_time', 'start_station_id']).user_type.value_counts().unstack(fill_value=0)
```

```python
# df of groups, trip duration average, genders, user_types
df_bike_transform = df_group_trip_duration.join(df_group_genders, on=['start_time', 'start_station_id']).join(df_group_user_type, on=['start_time', 'start_station_id'])
df_bike_transform.reset_index(inplace=True)
df_bike_transform.head()
```

## Feature Engineering

Because hour and day_of_week is a cyclical feature, we will encode it using cosine and sine

```python
df_fe = df_combined
```

```python
# Encode time series with cyclical features
df_fe['hour_sin'] = np.sin((2 * np.pi * df_fe['hour']) / max(df_fe['hour']))
df_fe['hour_cos'] = np.cos((2 * np.pi * df_fe['hour']) / max(df_fe['hour']))
df_fe['day_of_week_sin'] = np.sin((2 * np.pi * df_fe['day_of_week']) / max(df_fe['day_of_week']))
df_fe['day_of_week_cos'] = np.cos((2 * np.pi * df_fe['day_of_week']) / max(df_fe['day_of_week']))
df_fe['day_of_month_sin'] = np.sin((2 * np.pi * df_fe['day_of_month']) / max(df_fe['day_of_month']))
df_fe['day_of_month_cos'] = np.cos((2 * np.pi * df_fe['day_of_month']) / max(df_fe['day_of_month']))
df_fe.head()
```

```python
# Extract month and day from start_datetime
df_fe['month'] = df_fe['start_datetime'].dt.month
df_fe.head()
```

```python
df_fe.columns
```

```python
# Add day/night hours
df_fe['is_am_peak'] = (df_fe['hour']>=7) & (df_fe['hour']<=9)
df_fe['is_pm_peak'] = (df_fe['hour']>=16) & (df_fe['hour']<=19)
df_fe.head()
```
